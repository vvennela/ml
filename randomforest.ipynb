```python
from sklearn.ensemble import RandomForestRegressor

from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np

from sklearn import tree
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets
data = datasets.load_wine()
```

```python
print(data.DESCR)
```

```python
train_features, test_features, train_labels, test_labels = train_test_split(data.data, data.target, test_size = 0.25, random_state = 42)
```

```python
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python
# using the random forest regressor algorithm  
model = RandomForestRegressor(n_estimators = 100, 
                              oob_score = True, 
                              max_depth = 10, 
                              verbose = 1)
```

```python
model.fit(train_features, train_labels);
```

```python
model.score(test_features, test_labels)
# to increase the model score, create # estimators and max depth 
```

```python
estimator = model.estimators_[0]
```

```python
plt.figure(figsize = (16,10), dpi = 80, facecolor = 'w', edgecolor = 'k')

tree.plot_tree(estimator, feature_names = data.feature_names, rounded = True, filled = True, fontsize = 10)

plt.show()
```
